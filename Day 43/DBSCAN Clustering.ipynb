```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors
```

```python
X,y = make_moons(n_samples=1000, noise=0.05, shuffle=True)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

### Applying KMeans clustering on this dataset

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(centers[:,0],centers[:,1],color='red')
plt.show()
```

We can observe that KMeans algorithm can still fail in those datasets where the clusters formed are not linearlly separable or are complex in nature. They cannot easily assign clusters to appropriate classes.

## DBSCAN algorithm
 
  - **(Density Based Spatial Clustering of Applications with Noise)**
  - It is based upon the idea that a cluster is a high density area surrounded by low density regions.
  - It starts by exploring a small area if density is "good enough" it considers it as part of cluster and, explores neighbours goes so on, so basically it groups regions based upon density.

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

### Applying DBSCAN clustering on this dataset

```python
dbs = DBSCAN(eps=0.25,min_samples=5)
Y_pred = dbs.fit_predict(X)
print(Y_pred)
```

 - **eps** is the maximum distance between two samples for one to be considered  as in the neighborhood of the other.
 - **min_samples** is the number of samples (or total weight) in a neighborhood for a point to be considered as a core point. This includes the point itself.

```python
plt.scatter(X[:,0],X[:,1],c=Y_pred)
plt.show()
```

We can observe that DBSCAN is able to create clusters in a much better manner as compared to KMeans++ algorithm.
